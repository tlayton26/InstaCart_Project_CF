With User Defined Functions

```python
#create subset
df = ords_prods_merged[:1000000]
```

```python
#define variable
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#new column for price rnage
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#count
df['price_range'].value_counts(dropna = False )
```

```python
#agg
df['prices'].max()
```

Client wanting ranges for busiest time of day. 
Most orders between 10 and 16 
Fewest orders between 0 adn 6 
Everything else is in the average range. 

```python
#creating for-loop for busiest hours of day 
result_hrs = []

for value in ords_prods_merged["order_hour_of_day"]:
  if (value >= 10) & (value <= 16):
    result_hrs.append("Most Orders")
  elif (value >= 0) & (value <= 6):
    result_hrs.append("Fewest Orders")
  else:
    result_hrs.append("Average Orders")
```

```python
result_hrs
```

```python
# creating new column for busiest hours of day 
ords_prods_merged['busiest_period_of_day'] = result_hrs
```

```python
# Q5 checking frequency
ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
