```python
# steps 1 & 2. grouping by user id and aggregating max order number
ords_prods_der['max_order'] = ords_prods_der.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
#checking
ords_prods_der.head(15)
```

```python
#creating flag for loyalty customers 
ords_prods_der.loc[ords_prods_der['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_der.loc[(ords_prods_der['max_order'] <= 40) & (ords_prods_der['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_der.loc[ords_prods_der['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_der['loyalty_flag'].value_counts(dropna = False)
```

```python
#checking
ords_prods_der.head()
```

```python
# q5 multiple stats
ords_prods_der.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# grouping by cust_id and finding mean price spent 
ords_prods_der.groupby('customer_id').agg({'prices':['mean']})
```

```python
#creating column in df 
ords_prods_der['avg_spending'] = ords_prods_der.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
#check
ords_prods_der.head()
```

```python
#creating flag for spending 
ords_prods_der.loc[ords_prods_der['avg_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_der.loc[ords_prods_der['avg_spending'] > 10, 'spending_flag'] = 'High spender'
```

```python
pd.options.display.max_rows = None
```

```python
#check 
ords_prods_der.head(10)
```

```python
#count
ords_prods_der['spending_flag'].value_counts(dropna = False)
```

```python
# grouping by cust_id and finding median of days since prior prder
ords_prods_der.groupby('customer_id').agg({'days_since_prior_order':['median']})
```

```python
#creating column in df 
ords_prods_der['med_DSPO'] = ords_prods_der.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check 
ords_prods_der.head()
```

```python
#creating flag for freqency 
ords_prods_der.loc[ords_prods_der['med_DSPO'] > 20, 'freqency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_der.loc[(ords_prods_der['med_DSPO'] > 10) & (ords_prods_der['med_DSPO'] == 20), 'freqency_flag'] = 'Regular customer'
```

```python
ords_prods_der.loc[ords_prods_der['med_DSPO'] <= 10, 'freqency_flag'] = 'Frequent customer'
```

```python
ords_prods_der['freqency_flag'].value_counts(dropna = False)
```

```python
#oops. rename column
ords_prods_der.rename(columns = {'freqency_flag' : 'frequency_flag'}, inplace = True)
```
